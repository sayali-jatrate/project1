```
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1,2,figsize=(15,10))

ax=axes[0]
ax.set_xlim(0,10)
ax.set_ylim(0,10)
ax.axis('off')

circle=plt.Circle((5,5),1.2,color='red',alpha=0.85,zorder=5)
ax.add_patch(circle)
ax.text(5,5,'',ha='center',va='center',fontweight='bold',color='white',zorder=6)

inputs=[('x1',2.0,7.5,'w1'),('x2',2.0,5.0,'w2'),('x3',2.0,2.5,'w3')]
for name, x, y, z in inputs:
  ax.annotate('',xy=(3.8,5),xytext=(x+0.4,y),
              arrowprops=dict(arrowstyle='->',color='blue',lw=2))

  ax.text(x,y,name,ha='center',fontsize=14,fontweight='bold',color='gray',bbox=dict(boxstyle='round,pad=0.3',facecolor='#AEC6E8',alpha=0.8))
  ax.text((x+3.8)/2+0.1,(y+5)/2+0.2,z,fontsize=10,color='blue',fontweight='bold')

ax.text(4.2,3.4,'+b',fontsize=12,color='black',fontweight='bold')

ax.annotate('',xy=(8.5,5),xytext=(6.2,5),arrowprops=dict(arrowstyle='->',color='#2c3e50',lw=5))
ax.text(9.1,5,'y',ha='center',va='center',fontsize=16,fontweight='bold',color='#55A868',bbox=dict(boxstyle='round,pad=0.1',facecolor='#A3D5B5',alpha=0.0))

ax.text(5,9.3,'Single Artificial Neuron',ha='center',fontsize=13,fontweight='bold',color='pink')

ax.text(5,8.7,'z=w1x1+w2x2+w3x3+b\ny=o(z)',ha='center',fontsize=10,color='#555',bbox=dict(boxstyle='round',facecolor='#f9f9f9',alpha=0.9))

ax2=axes[1]
z_arr=np.linspace(-5,5,300)

activations={
    'sigmoid o(z)=1/(1+np.exp(-z))':(1/(1+np.exp(-z_arr)),'red'),
    'Tanh tanh(z)':(np.tanh(z_arr),'blue'),
    'ReLU max(0,z):':(np.maximum(0,z_arr),'green'),
    'Leaky ReLU':(np.where(z_arr>0,z_arr,0.1*z_arr),'pink')
}
for name, (vals,color) in activations.items():
  ax2.plot(z_arr,vals,lw=2.5,color=color,label=name)

  ax2.axhline(0,color='black',lw=0.8,alpha=0.5)
  ax2.axvline(0,color='black',lw=0.8,alpha=0.5)
  ax2.set_xlabel('z(pre-activation)',fontsize=11)
  ax2.set_ylabel('a(activation)',fontsize=11)
  ax2.set_title('Activation function')
  ax2.legend(fontsize=9)
  ax2.set_ylim(-1.5,1.5)
  ax2.annotate('ReLU:Most common\n in hidden layers',
               xy=(2.5,2.5),xytext=(1,-1.2),arrowprops=dict(arrowstyle='->',color='black'),
               fontsize=8,color='blue')
plt.suptitle('Deep Learning Fundamentals- The Artificial Neuron',fontsize=13,fontweight='bold')
plt.tight_layout()
plt.show()
```
